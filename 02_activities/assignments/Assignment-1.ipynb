```python
#Import Libraries 
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#Sample data to get basic plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

# Determining size and type of graph as Scatter plot
fig, ax = plt.subplots(figsize = (5,3))
ax.scatter (x,y)
```

```python
#Change to Bar graph
fig, ax = plt.subplots(figsize = (5,3))
ax.bar (x,y)
```

```python
#Modifying type of graph to histogram
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(y)
#Adding title, X & Y Label 
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
#Creating dictionay for the labels
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

#Adding to previous plot 
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('Total Growth Over Time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
#Changing color and marker
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y,
           marker = '*',
           color = 'indigo')
```

```python
#Adding colour, marker and lines connecting the points
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,
        y,
        marker = '*',
        color = 'indigo',
        linestyle = '--',
        linewidth = '2')
```

```python
#Using hex codes to plot
fig, ax = plt.subplots(figsize= (5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')
# Adding y- axis
ax.grid(axis='y')
```

```python
fig, ax = plt.subplots(figsize= (5,3))
ax.plot(x,y,
        marker = '*',
        markersize = 12,
        color = '#7425b9',
        linestyle = '--',
        linewidth = 2,
        markeredgecolor = '#fa9359',
        markerfacecolor = '#000000')

# Modifying appearance of y-axis
ax.grid(color = 'pink', linewidth = 4, linestyle = 'dotted')
```

```python
#Activity 

import matplotlib.pyplot as plt
 
# Create bars
barWidth = 0.9
bars1 = [3, 3, 1]
bars2 = [4, 2, 3]
bars3 = [4, 6, 7, 10, 4, 4]
bars4 = bars1 + bars2 + bars3
 
# The X position of bars
r1 = [1,5,9]
r2 = [2,6,10]
r3 = [3,4,7,8,11,12]
r4 = r1 + r2 + r3
 
# Create barplot
plt.bar(r1, bars1, width = barWidth, color = (0.3,0.1,0.4,0.6), label='Alone')
plt.bar(r2, bars2, width = barWidth, color = (0.3,0.5,0.4,0.6), label='With Himself')
plt.bar(r3, bars3, width = barWidth, color = (0.3,0.9,0.4,0.6), label='With other genotype')
# Note: the barplot could be created easily. See the barplot section for other examples.
 
# Create legend
plt.legend()
 
# Text below each barplot with a rotation at 90°
plt.xticks([r + barWidth for r in range(len(r4))], ['DD', 'with himself', 'with DC', 'with Silur', 'DC', 'with himself', 'with DD', 'with Silur', 'Silur', 'with himself', 'with DD', 'with DC'], rotation=90)
 
# Create labels
label = ['n = 6', 'n = 25', 'n = 13', 'n = 36', 'n = 30', 'n = 11', 'n = 16', 'n = 37', 'n = 14', 'n = 4', 'n = 31', 'n = 34']
 
# Text on the top of each bar
for i in range(len(r4)):
    plt.text(x = r4[i]-0.5 , y = bars4[i]+0.1, s = label[i], size = 6)

# Adjust the margins
plt.subplots_adjust(bottom= 0.2, top = 0.98)
 
# Show graphic
plt.show()
```
